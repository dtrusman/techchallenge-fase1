## **Síndrome da Fadiga Crônica (EM/SFC) e Depressão**

A **Síndrome de Fadiga Crônica**, também conhecida como  **Encefalomielite Miálgica,** é caracterizada por deixar o portador com uma sensação de estar extremamente cansado o tempo todo, mesmo após uma boa noite de sono. Esta condição pode piorar após qualquer tipo de atividade, exercício físico ou mental. A pessoa fica exausta para fazer quase todas as suas atividades diárias, como: tomar banho, ir ao trabalho, fazer serviços domésticos e demais atividades que necessitem sair de casa.

Além da fadiga, a doença pode gerar sintomas como: dificuldade para dormir ou acordar, tontura, fraqueza muscular, dor de garganta, problemas digestivos, calafrios, suores noturnos, alergias ou reações alimentares, sensibilidade à luz ou à ruídos, batimentos cardíacos irregulares e falta de ar.

A **Depressão** é uma doença de humor, que pode causar tristeza profunda ou apatia constante, em que o indivíduo não sente mais prazer ou sentido na vida, com sentimos de culpa, vazio e inutilidade. Em casos mais graves, o portador da doença pode pensar ou realmente atentar contra a própria vida.

Na depressão, sintomas como fadiga também podem estar presentes, assim como, problemas de sono (dormir muito ou pouco), falta de energia para realizar tarefas simples,  ansiedade, lapsos de memórias ou concentração, fala ou movimentos lentos, dor sem causa física, problemas digestivos ou cólicas abdominais e inquietação.

As duas doenças são crônicas e podem durar anos, com períodos de melhora e outros de piora do quadro. Ambas as condições tem sintomas em comum, tais como: fadiga, problemas de sono, falta de energia, dor, problemas digestivos, dores de cabeça, lapsos de memória ou dificuldade de concentração. A depender dos sintomas, pode haver dificuldade do diagnóstico correto para o paciente, que pode apresentar uma das comorbidades ou ambas.

O trabalho a seguir tem como objetivo responder à seguinte questão: é possível treinar um modelo de machine learning capaz de classificar pacientes em uma das três categorias — depressão, fadiga crônica ou ambas — a partir de variáveis clínicas?

Este é o primeiro conjunto de dados desse tipo criado especificamente para ajudar iniciantes na área e pesquisadores a explorar casos complexos de diagnóstico em condições mentais e crônicas de saúde.



# **Explicação das variáveis presentes nos dados:**

* **age** -	Idade do paciente:
    * Variando de 18 à 70 anos.
* **gender** -	Gênero:
    * Feminio,
    * Masculino.
* **fatigue_severity_scale_score** -	Escala de Gravidade da Fadiga, é uma das ferramentas de avaliação da fadiga mais comumente usadas na prática clínica e na pesquisa, funcionando como  um modo de avaliar o impacto da fadiga no paciente. O teste é baseado em um questionário com 9 perguntas, que devem ser respondidas com os números de 1 à 7 , que representam respectivamente, completamente em desacordo com a sentença (1) e completamente em acordo (7).
Alguns estudos utilizam a média das respostas, e outros a soma total das respostas (variando de 9 a 63). Uma pontuação inferior a 36 sugere que o paciente pode não estar sofrendo de fadiga e um valor superior sugere a presença de fadiga. No caso da escala apresentada neste banco de dados, estes estão normalizados.
    * Escala de 1 à 10.
* **depression_phq9_score** -	(PHQ-9) Índice de depressão, questionário que avalia a gravidade de depressão do paciente em que as questões somam um valor de 1 à 27, e classificam a depressão como a seguir:
    * 1-4 - Depressão mínima
    * 5-9 - Depressão leve
    * 10-14 - Depressão moderada
    * 15-19 - Depressão moderadamente grave
    * 19-27 - Depressão grave
* **pem_present	Whether** - Mal-estar pós-esforço está presente, questiona se o paciente em questão sofreu algum tipo de mal estar após algum tipo de esforço físico.
    * Sim,
    * Não.
* **pem_duration_hours** -	Duração do mal-estar pós-esforço (PEM) em horas;
* **sleep_quality_index** -	Qualidade do sono ((aumento da qualidade de sono do paciente de acordo com o aumento da escala);
    * Escala de 1 à 10.
* **brain_fog_level** -	Confusão mental / Dificuldade de concentração ((aumento da confusão mental do paciente de acordo com o aumento da escala);
    * Escala de 1 à 10.
* **physical_pain_score** -	Intensidade de dor física (aumento da dor física do paciente de acordo com o aumento da escala);
    * Escala de 1 à 10.
* **stress_level** - Nível de estresse (aumento do nível de estresse do paciente de acordo com o aumento da escala)
    * Escala de 1 à 10.
* **work_status** -	Status de trabalho:
    * Trabalhando,
    * Trabalho parcial,
    * Sem trabalho.
* **social_activity_level** - Nível de atividade social:
    * Muito alto,
    * Alto,
    * Médio,
    * Baixo,
    * Muito baixo.
* **exercise_frequency** - Frequência de atividade física:
    * Diariamente,
    * Frequentemente,
    * Às vezes,
    * Raramente,
    * Nunca.
* **meditation_or_mindfulness** - O paciente pratica mindfulness ou meditação?  
    * Sim,
    * Não.
* **hours_of_sleep_per_night** - Duração média de sono por noite
* **diagnosis** - (Diagnóstico) Qual(is) doença(s) o paciente tem? **(Variável Target)**
    * ME/CFS (síndrome de fadiga crônica),
    * Depressão,
    * Ambas (síndrome de fadiga crônica e depressão.

Vamos analisar todas as variáveis numéricas à procura de **outliers** utilizando os gráficos de boxplot:

```python
fig, axes = plt.subplots(3, 3, figsize=(18, 15)) # Create a 3x3 grid of subplots

# Flatten the axes array for easy iteration
axes = axes.flatten()

# List of columns to plot
columns_to_plot = [
    "age",
    "sleep_quality_index",
    "brain_fog_level",
    "physical_pain_score",
    "depression_phq9_score",
    "stress_level",
    "hours_of_sleep_per_night",
    "pem_duration_hours",
    "fatigue_severity_scale_score"
]

# Titles for the plots
plot_titles = [
    "Idade",
    "Qualidade do sono",
    "Confusão mental / Dificuldade de concentração",
    "Intensidade da dor física",
    "Índice de depressão - PHQ9",
    "Nível de estresse",
    "Duração média do sono por noite",
    "Duração do mal-estar pós-esforço (PEM) em horas",
    "Escala de Gravidade da Fadiga"
]

# Iterate through columns and axes to create box plots
for i, col in enumerate(columns_to_plot):
    sns.boxplot(y=data[col], ax=axes[i])
    axes[i].set_title(plot_titles[i])
    axes[i].set_ylabel("") # Remove y-axis label to avoid clutter

plt.tight_layout()
plt.show()
```

Analisando os gráficos, vemos que o "Índice de depressão (PHQ9)" apresenta um outlier, acima do valor de 25 e abaixo do valor de 19. Já o gráfico para a variável "Escala de gravidade da fadiga", há vários outliers, que estão acima ou abaixo da mediana.
Em ambos os casos, embora estatisticamente os dados sejam considerados como outliers, pois, estão longe da concentração central (mediana), não devemos desconsiderá-los para a predição do modelo.
Os dados, conforme a literatura, estão dentro do valor previsto para cada caso. No caso do Índice de depressão, este pode variar de 0 à 29, e podemos ver pelo gráfico que o outlier está dentro deste valor. O mesmo acontece para a Escala de gravidade da fadiga, todos os valores apresentados como outliers, estão dentro da escala esperada, entre 0 e 10. O que acontece neste caso é que, pode haver uma alta concentração de dados em torno da mediana e os dados podem estar assimetricamente distribuídos. Isto pode ser visto no gráfico a seguir, para o caso da Escala de gravidade da fadiga:

```python
coluna = data["fatigue_severity_scale_score"].dropna()  # remove NaNs, se houver

mediana = np.median(coluna)

plt.figure(figsize=(8, 4))
sns.histplot(coluna, bins=10, kde=True, color="skyblue")

# Linha vertical indicando a mediana
plt.axvline(mediana, color='red', linestyle='--', linewidth=2, label=f'Mediana = {mediana:.2f}')

plt.text(mediana + 0.1, plt.ylim()[1]*0.9, f'Mediana\n{mediana:.2f}', color='red')

plt.title("Histograma: Escala de gravidade da fadiga e respectiva mediana")
plt.xlabel("Valor")
plt.ylabel("Frequência")
plt.legend()
plt.show()
```

Observamos que alguns apresentaram acurácias muito elevadas (entre 0,92 e 0,99), enquanto outros, como o K-Nearest Neighbors, obtiveram desempenho significativamente inferior (0,58). É importante ressaltar que, embora modelos com alta acurácia sejam atrativos, eles devem ser interpretados com cautela, pois existe a possibilidade de que tenham apenas "decorado" os padrões do conjunto de treino, o que reduziria sua capacidade de generalização e, consequentemente, sua utilidade em análises futuras.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

classifiers = [
    ('Decision Tree', DecisionTreeClassifier(random_state=42)),
    ('Random Forest', RandomForestClassifier(random_state=42)),
    ('AdaBoost', AdaBoostClassifier(random_state=42)),
    ('Gradient Boosting', GradientBoostingClassifier(random_state=42)),
    ('K-Nearest Neighbors', KNeighborsClassifier()),
    ('Naive Bayes', GaussianNB())
]

for model_name, clf in classifiers:
    pipe = Pipeline([
        ("scaler", StandardScaler()),
        ("model", clf)
    ])
    pipe.fit(x_train, y_train)
    y_pred = pipe.predict(x_test)
    print(f"Acurácia do modelo {model_name}: {accuracy_score(y_test, y_pred):.4f}")
```

## **Diagnóstico de overfitting: treino vs teste**

Para distinguir overfitting de vazamento de rótulo, comparamos as métricas no conjunto de **treino** e no de **teste**. Overfitting clássico se manifesta como `acc_treino >> acc_teste`.

```python
print("=== Desempenho no TREINO ===")
print(classification_report(y_train, pipe_lr.predict(x_train)))
print("\n=== Desempenho no TESTE ===")
print(classification_report(y_test, predictions_lr))
```

## **Experimento ablativo (Remoção): identificando o vazamento de rótulo**

Features como `pem_present`, `depression_phq9_score` e `fatigue_severity_scale_score` são instrumentos clínicos que **definem ou quantificam** as doenças diagnosticadas. Usá-las como preditoras cria circularidade: o modelo aprende o diagnóstico a partir do próprio diagnóstico.

Removemos essas features e re-treinamos o modelo para quantificar o impacto.

```python
features_diagnosticas = ["pem_present", "pem_duration_hours", "depression_phq9_score", "fatigue_severity_scale_score"]
x_sem_diag = x.drop(features_diagnosticas, axis=1)

x_train_nd, x_test_nd, y_train_nd, y_test_nd = train_test_split(
    x_sem_diag, y, test_size=0.2, random_state=42, stratify=y
)

pipe_ablativo = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(class_weight="balanced", max_iter=2000))
])
pipe_ablativo.fit(x_train_nd, y_train_nd)
pred_sem_diag = pipe_ablativo.predict(x_test_nd)

acc_original = accuracy_score(y_test, predictions_lr)
acc_sem_diag = accuracy_score(y_test_nd, pred_sem_diag)

print(f"Acurácia COM todas as features (modelo original): {acc_original:.3f}")
print(f"Acurácia SEM features diagnósticas:               {acc_sem_diag:.3f}")
print(f"Queda na acurácia:                                 {acc_original - acc_sem_diag:.3f}")
print("\n=== Relatório SEM features diagnósticas ===")
print(classification_report(y_test_nd, pred_sem_diag))
```

## **Validação cruzada estratificada (5-fold)**

Um único split treino/teste pode ser influenciado pelo acaso da divisão. A validação cruzada com 5 folds produz uma estimativa mais robusta, comparando ambos os modelos (com e sem features diagnósticas).

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

cv_scores_full = cross_val_score(pipe_lr, x, y, cv=cv, scoring='f1_macro')
cv_scores_nd   = cross_val_score(pipe_ablativo, x_sem_diag, y, cv=cv, scoring='f1_macro')

print(f"F1-macro com TODAS as features (5-fold): {cv_scores_full.mean():.3f} ± {cv_scores_full.std():.3f}")
print(f"F1-macro SEM features diagnósticas:      {cv_scores_nd.mean():.3f} ± {cv_scores_nd.std():.3f}")

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(range(1, 6), cv_scores_full, 'o-', label='Todas as features', color='steelblue')
ax.plot(range(1, 6), cv_scores_nd,   'o-', label='Sem features diagnósticas', color='tomato')
ax.set_xlabel('Fold')
ax.set_ylabel('F1-macro')
ax.set_title('Validação Cruzada: Todas as features vs Sem diagnóstico')
ax.set_ylim(0, 1.05)
ax.legend()
plt.tight_layout()
plt.show()
```

## **Curvas ROC e AUC multiclasse**

A curva ROC e a métrica AUC mostram a capacidade discriminativa do modelo para cada classe, independente do limiar de decisão. Usamos a abordagem One-vs-Rest (OvR).

```python
classes_labels = sorted(y.unique())

y_score    = pipe_lr.predict_proba(x_test)
y_test_bin = label_binarize(y_test, classes=classes_labels)

roc_auc_macro = roc_auc_score(y_test, y_score, multi_class='ovr', average='macro')
print(f"ROC-AUC macro (Regressão Logística, todas as features): {roc_auc_macro:.3f}")

fig, ax = plt.subplots(figsize=(8, 6))
cores = ['royalblue', 'forestgreen', 'tomato']
for i, (color, cls) in enumerate(zip(cores, classes_labels)):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_score[:, i])
    roc_auc_val  = auc(fpr, tpr)
    ax.plot(fpr, tpr, color=color, lw=2, label=f'{cls} (AUC = {roc_auc_val:.2f})')

ax.plot([0, 1], [0, 1], 'k--', lw=1, label='Classificador aleatório')
ax.set_xlabel('Taxa de Falsos Positivos')
ax.set_ylabel('Taxa de Verdadeiros Positivos (Recall)')
ax.set_title('Curvas ROC — Regressão Logística (One-vs-Rest)')
ax.legend(loc='lower right')
plt.tight_layout()
plt.show()
```

# **Conclusão**

Neste estudo, desenvolvemos modelos de inteligência artificial para classificar pacientes entre três condições: **Depressão**, **ME/CFS (Fadiga Crônica)** e **Ambas as doenças**, com um dataset sintético de 1000 pacientes e 16 variáveis.

## A acurácia de ~99%: overfitting ou vazamento de rótulo?

A acurácia elevada observada nos modelos baseados em árvore **não é overfitting clássico**. Overfitting ocorre quando `acc_treino >> acc_teste` — o modelo memoriza o treino e falha no teste. Como demonstrado na seção "Diagnóstico de overfitting", as métricas no treino e no teste são comparáveis, descartando essa hipótese.

O verdadeiro problema, confirmado pelo **experimento ablativo (remoção)**, é o **vazamento de rótulo (label leakage)**: features como `pem_present`, `depression_phq9_score` e `fatigue_severity_scale_score` são instrumentos clínicos que definem as próprias doenças que o modelo prevê. Ao removê-las, a acurácia cai substancialmente — revelando que os modelos aprenderam a explorar essa circularidade em vez de padrões genuinamente preditivos. O gráfico de importância de features confirma visualmente que essas variáveis dominam a predição.

## Limitações

- **Dataset sintético com padrões determinísticos:** classes separadas artificialmente, sem ruído clínico real;
- **Features circulares:** PHQ-9 e escala de fadiga pressupõem o diagnóstico;
- **Desbalanceamento de classes:** a classe "Ambas" tem ~50% menos amostras (recall ~0,69 na Regressão Logística), prejudicando a robustez nessa classe.

## Próximos passos para um estudo clinicamente válido

1. Coletar dados clínicos **reais e prospectivos**, registrados antes do diagnóstico final;
2. Usar apenas features disponíveis **antes** do diagnóstico (demográficas, sono, atividade física, dor, estresse);
3. Aplicar balanceamento de classes (SMOTE) **somente no conjunto de treino**, via `imblearn.pipeline`;
4. Validar em um conjunto externo para medir generalização real;
5. Analisar interpretabilidade com SHAP values para identificar os fatores que realmente diferenciam as condições.
